# Introduction to Flight Price Prediction using Various Regression Models

## Overview
This project predicts flight prices using various regression models. The dataset includes features such as airline, source and destination cities, class, stops, and duration.

## Objective
Develop and compare regression models to predict flight prices based on given features.

## Steps
1. **Data Loading:** Import and preview the dataset.
2. **Data Preparation:** Clean and preprocess the data, including encoding and scaling.
3. **Model Training:** Train multiple regression models.
4. **Model Evaluation:** Evaluate models using metrics like MAE, MSE, RMSE, and R2 score.
5. **Result Visualization:** Visualize actual vs. predicted prices and other insights.

```python
# Importing train_test_split function
from sklearn.model_selection import train_test_split  # Imports the train_test_split function, which is used to randomly split the data into training and testing sets.

# Splitting the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)  # Splits the data into training and testing sets, with 30% of the data reserved for testing. The random_state parameter ensures reproducibility of the split.

# Getting the shapes of the training and testing sets
x_train.shape, x_test.shape, y_train.shape, y_test.shape  # Outputs the dimensions of the training and testing sets for both features (x) and target (y).
```

```python
# Importing MinMaxScaler
from sklearn.preprocessing import MinMaxScaler  # Imports the MinMaxScaler, which is used to scale features to a given range, typically 0 to 1.

# Initializing the MinMaxScaler with a feature range of 0 to 1
mmscaler = MinMaxScaler(feature_range=(0, 1))  # Creates an instance of MinMaxScaler, setting the feature range to scale between 0 and 1.

# Scaling the training data
x_train = mmscaler.fit_transform(x_train)  # Fits the scaler to the training data and then transforms it. This means the scaler adjusts based on the training data's properties and applies the scaling.

# Scaling the testing data
x_test = mmscaler.fit_transform(x_test)  # Fits the scaler to the test data and then transforms it. Note: This is not typically recommended as it should be fit only once on training data.

# Converting scaled arrays back to DataFrame format
x_train = pd.DataFrame(x_train)  # Converts the scaled numpy array back into a DataFrame for ease of use in further analysis.
x_test = pd.DataFrame(x_test)  # Converts the scaled numpy array back into a DataFrame.
```

```python
# Initializing an empty dictionary for storing model results
a = {'Model Name': [], 'Mean_Absolute_Error_MAE': [], 'Adj_R_Square': [], 'Root_Mean_Squared_Error_RMSE': [],
     'Mean_Absolute_Percentage_Error_MAPE': [], 'Mean_Squared_Error_MSE': [], 'Root_Mean_Squared_Log_Error_RMSLE': [],
     'R2_score': []}

# Creating a DataFrame from the dictionary
Results = pd.DataFrame(a)  # Converts the dictionary 'a' into a DataFrame. Since the lists are empty, this DataFrame will also be empty initially.

# Displaying the first few rows of the DataFrame
Results.head()  # Shows the first five rows of the DataFrame 'Results'. Since it's empty, it will display an empty DataFrame with the columns specified.
```

```python
# Importing regression models from scikit-learn and other libraries
from sklearn import metrics  # Importing metrics module for evaluation metrics

# Linear models
from sklearn.linear_model import LinearRegression  # Imports Linear Regression, a basic but widely used regression model.
from sklearn.linear_model import Ridge  # Imports Ridge Regression, a type of linear regression with L2 regularization.
from sklearn import linear_model  # Imports the linear_model module from scikit-learn for other linear models.

# Decision Tree and Ensemble models
from sklearn.tree import DecisionTreeRegressor  # Imports Decision Tree Regressor, a model that predicts target variables by learning decision rules from features.
from sklearn.ensemble import RandomForestRegressor  # Imports Random Forest Regressor, an ensemble method that uses multiple decision trees to improve predictive performance.
from sklearn.ensemble import ExtraTreesRegressor  # Imports Extra Trees Regressor, another ensemble method similar to Random Forest but with some variations in how trees are built.
from sklearn.ensemble import BaggingRegressor  # Imports Bagging Regressor, an ensemble method that improves the stability and accuracy of machine learning algorithms.

# Support Vector Machine
from sklearn.svm import SVR  # Imports Support Vector Regressor, which is based on the support vector machine algorithm.

# Gradient Boosting
import xgboost as xgb  # Imports XGBoost, a powerful gradient boosting framework optimized for performance and speed.

# Nearest Neighbors
from sklearn.neighbors import KNeighborsRegressor  # Imports K-Neighbors Regressor, a model that predicts the target variable based on the average of the nearest neighbors.

# Displaying the head of the results DataFrame
Results.head()  # Displays the first few rows of the initialized results DataFrame to confirm its structure.
```

```python
# Create objects of Regression / Regressor models with default hyper-parameters

modelmlg = LinearRegression()  # Creates an instance of Linear Regression model with default parameters.
modeldcr = DecisionTreeRegressor()  # Creates an instance of Decision Tree Regressor with default parameters.
modelbag = BaggingRegressor()  # Creates an instance of Bagging Regressor with default parameters.
modelrfr = RandomForestRegressor()  # Creates an instance of Random Forest Regressor with default parameters.
modelSVR = SVR()  # Creates an instance of Support Vector Regressor with default parameters.
modelXGR = xgb.XGBRegressor()  # Creates an instance of XGBoost Regressor with default parameters.
modelKNN = KNeighborsRegressor(n_neighbors=5)  # Creates an instance of K-Neighbors Regressor with default number of neighbors set to 5.
modelETR = ExtraTreesRegressor()  # Creates an instance of Extra Trees Regressor with default parameters.
modelRE = Ridge()  # Creates an instance of Ridge Regression model with default parameters.
modelLO = linear_model.Lasso(alpha=0.1)  # Creates an instance of Lasso Regression model with alpha (regularization strength) set to 0.1.
```

```python




# Evalution matrix for all the algorithms

MM = [modelmlg, modeldcr, modelrfr, modelKNN, modelETR, modelXGR, modelbag,modelRE,modelLO]

# Assuming that you have already initialized the `Results` DataFrame as shown previously

for models in MM:
    models.fit(x_train, y_train)
    y_pred = models.predict(x_test)

    # Evaluation metrics
    mae = round(metrics.mean_absolute_error(y_test, y_pred), 3)
    mse = round(metrics.mean_squared_error(y_test, y_pred), 3)
    rmse = round(np.sqrt(mse), 3)
    r2 = round(metrics.r2_score(y_test, y_pred), 6)
    rmsle = round(np.log(rmse), 3)

    def MAPE(y_test, y_pred):
        y_test, y_pred = np.array(y_test), np.array(y_pred)
        return np.mean(np.abs((y_test - y_pred) / y_test)) * 100

    result_mape = MAPE(y_test, y_pred)

    r_squared = r2
    adjusted_r_squared = round(1 - (1-r_squared)*(len(y)-1)/(len(y)-x.shape[1]-1), 6)

    # Create a dictionary for the new row
    new_row = {'Model Name': type(models).__name__, 'Mean_Absolute_Error_MAE': mae,
               'Adj_R_Square': adjusted_r_squared, 'Root_Mean_Squared_Error_RMSE': rmse,
               'Mean_Absolute_Percentage_Error_MAPE': result_mape,
               'Mean_Squared_Error_MSE': mse, 'Root_Mean_Squared_Log_Error_RMSLE': rmsle,
               'R2_score': r2}

    # Append new row to Results DataFrame using pd.concat
    Results = pd.concat([Results, pd.DataFrame([new_row])], ignore_index=True)

    print('------------------------------------------------------------------------------------------------------------')
```
